```python
# SIMULATION PARAMETERS
beta, mu = 33.7, 96*np.pi/180 # Twiss parameters
M = np.array([[np.cos(mu), beta * np.sin(mu)],
             [-np.sin(mu)/beta, np.cos(mu)]]) # Transfer matrix
x1, x2 = np.array([4,0]), np.array([0,0.3]) # mm, mrad
print(M)
```

```python
# PARTICLE TRACK FUNCTION
def get_trajectory(x0, lattice_elements, M):
    trajectory = [x0]
    for _ in range(lattice_elements):
        x0 = M @ x0
        trajectory.append(x0)
    return np.array(trajectory)
```

```python
# GET TRAJECTORIES FOR BOTH PARTICLES
trajectory1 = get_trajectory(x1, 24, M)
trajectory2 = get_trajectory(x2, 24, M)
```

```python
# PLOT TRAJECTORIES
plt.figure(figsize=(10, 8))
plt.plot(trajectory1[:,0], trajectory1[:,1], color='blue', alpha=0.7)
plt.plot(trajectory2[:,0], trajectory2[:,1], color='green', alpha=0.7)
plt.plot(x1[0], x1[1], color='blue', marker='*', markersize=15, label='Start 1')
plt.plot(x2[0], x2[1], color='green', marker='*', markersize=15, label='Start 2')
plt.xlabel('x (mm)')
plt.ylabel("x' (mrad)")
plt.legend()
```

```python
# PARTICLE TRACK FUNCTION
def get_trajectory_with_RMS(x0, lattice_elements, M):
    trajectory = [x0]
    x0 = M @ x0
    x0[1] += np.random.normal(0, 0.106)
    trajectory.append(x0)
    for _ in range(lattice_elements-1):
        x0 = M @ x0
        trajectory.append(x0)
    return np.array(trajectory)
```

```python
# GET TRAJECTORIES FOR BOTH PARTICLES
trajectory1 = get_trajectory_with_RMS(x1, 24, M)
trajectory2 = get_trajectory_with_RMS(x2, 24, M)
```

```python
# PLOT TRAJECTORIES
plt.figure(figsize=(10, 8))
plt.plot(trajectory1[:,0], trajectory1[:,1], color='blue', alpha=0.7)
plt.plot(trajectory2[:,0], trajectory2[:,1], color='green', alpha=0.7)
plt.plot(x1[0], x1[1], color='blue', marker='*', markersize=15, label='Start 1')
plt.plot(x2[0], x2[1], color='green', marker='*', markersize=15, label='Start 2')
plt.xlabel('x (mm)')
plt.ylabel("x' (mrad)")
plt.legend()
```
