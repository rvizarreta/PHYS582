```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calculate_ratio(m0, E): # Energy in MeV
    f_100 = (0.1+m0)/m0
    gamma_100 = (0.1+m0)/m0
    f_E = (E+m0)/m0
    gamma_E = (E+m0)/m0
    res = np.sqrt(1-1/gamma_E**2)/np.sqrt(1-1/gamma_100**2)
    return res
```

```python
KE_range = np.arange(0,1e3,10)
```

```python
plt.plot(KE_range, calculate_ratio(938, KE_range), label='Proton')
plt.plot(KE_range, calculate_ratio(1875, KE_range), label='Deuteron')
plt.plot(KE_range, calculate_ratio(0.511, KE_range), label='Electron')
plt.xlabel('KE (MeV)')
plt.ylabel(r'$f_{E}/f_{100 keV}$')
plt.legend()
```
