```python
N_array, iterations = np.arange(0,100,5), 10000
variance_array = []

for N in N_array:
    end_position = []
    for iteration in range(iterations):
        position = 0
        throws = np.random.randint(0, 2, size=int(N))
        for throw in throws:
            if throw == 0: #tails
                position -= 1
            else: #heads
                position += 1
        end_position.append(position)

    end_position = np.array(end_position)
    variance_array.append(np.var(end_position))
    
x_range = np.linspace(0, 100, 100)
plt.plot(x_range, np.sqrt(x_range), label=r'$\sqrt{N}$', color='red', linestyle='-')
plt.scatter(N_array, np.sqrt(variance_array))
plt.xlabel('N')
plt.ylabel(r'$\sqrt{VAR}$')
plt.legend()
```

```python
# Random number generation following exponential distribution
def generate_random_dist_exp(size):
    F_inv = lambda u, b : -b*np.log(1-u)
    uniform_dist = np.random.uniform(low=0.0, high=1.0, size=size)
    x_random_values = F_inv(uniform_dist, 1)
    return x_random_values

plot_manager.display_plots(plot_funcs = {
    'Plot 1': lambda: plot_manager.make_hist(
        data=[generate_random_dist_exp(10000)], 
        bins=60, 
        range=(0, 10), 
        alpha=0.5, 
        label=[r'$\pi$'], 
        title=r'PDF', 
        xlabel='x', 
        ylabel='No. of counts',
        legend=False,
        DUNE=False
    )
})
```

```python
N_array, iterations = np.arange(1,100,5), 10000
variance_array = []

for N in N_array:
    end_position = []
    for iteration in range(iterations):
        position = 0
        throws = np.random.randint(0, 2, size=int(N))
        step_sizes = generate_random_dist_exp(int(N))
        for throw, step_size in zip(throws,step_sizes):
            if throw == 0: #tails
                position -= step_size
            else: #heads
                position += step_size
        end_position.append(position)

    end_position = np.array(end_position)
    variance_array.append(np.var(end_position))
    
x_range = np.linspace(0, 100, 100)
plt.plot(x_range, (7/5)*np.sqrt(x_range), label=r'$\frac{7}{5}\sqrt{N}$', color='red', linestyle='-')
plt.scatter(N_array, np.sqrt(variance_array))
plt.xlabel('N')
plt.ylabel(r'$\sqrt{VAR}$')
plt.title('1D Mean Free Path')
plt.legend()
```
