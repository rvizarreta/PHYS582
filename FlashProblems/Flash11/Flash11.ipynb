```python
calculate_Mv = lambda r : np.sqrt(12/r**2) # r in fm
G2 = lambda Q, r : 1/(1+Q**2/(calculate_Mv(r))**2)**2
A = lambda Q, M, r : (G2(Q,r) + (Q**2/(4*M**2))*G2(Q,r))/(1 + Q**2/(4*M**2))
B = lambda Q, M, r : 2*(Q**2/(4*M**2))*G2(Q,r)
```

```python
r1, r2, r3, r4, r5 = 0.2, 0.5, 0.8, 1, 2 # fm
```

```python
M = 0.938 # GeV
q_range = np.arange(0,10,0.01)
```

```python
plt.plot(q_range, A(-q_range, M, r1), label='0.2 fm')
plt.plot(q_range, A(-q_range, M, r2), label='0.5 fm')
plt.plot(q_range, A(-q_range, M, r3), label='0.8 fm')
plt.plot(q_range, A(-q_range, M, r4), label='1 fm')
plt.plot(q_range, A(-q_range, M, r5), label='2 fm')
plt.xlabel(f'$q^2$ $(GeV^2)$')
plt.ylabel(f'$A(Q^2)$')
plt.legend()
```

```python
plt.plot(q_range, B(-q_range, M, r1), label='0.2 fm')
plt.plot(q_range, B(-q_range, M, r2), label='0.5 fm')
plt.plot(q_range, B(-q_range, M, r3), label='0.8 fm')
plt.plot(q_range, B(-q_range, M, r4), label='1 fm')
plt.plot(q_range, B(-q_range, M, r5), label='2 fm')
plt.xlabel(f'$q^2$ $(GeV^2)$')
plt.ylabel(f'$B(Q^2)$')
plt.legend()
```
