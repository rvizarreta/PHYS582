```python
from PionDecaySim import PionDecaySim
import numpy as np
```

```python
decay = PionDecaySim(10000, 1000.0, 100.0)
pions = decay.pion_decay_lab_frame()
muon_momentum_rest, nu_momentum_rest = decay.after_decay_particles_rest_frame()
```

```python
muons_lab, neutrinos_lab = [], []
for i in range(decay.n_particles):
    muon = decay.after_decay_particles_lab_frame(muon_momentum_rest[i], 
                                 decay.MUON_MASS, 
                                 pions[i])
    nu = decay.after_decay_particles_lab_frame(nu_momentum_rest[i],
                               0,
                               pions[i])
    muons_lab.append(muon)
    neutrinos_lab.append(nu)
```

```python
def analyze_results(muons):
    """Analyze and plot results of the simulation"""
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D
    
    # Extract muon momentum and energies
    momentum = np.array([np.linalg.norm(muon.momentum) for muon in muons])
    energies = np.array([muon.energy for muon in muons])
    angles = np.array([np.arccos(muon.momentum[1]/np.linalg.norm(muon.momentum)) 
                      for muon in muons])
    
    # Print summary statistics
    print(f"Muon momentum: mean = {np.mean(momentum):.2f} MeV/c, "
          f"std = {np.std(momentum):.2f} MeV/c")
    print(f"Muon energy: mean = {np.mean(energies):.2f} MeV, "
          f"std = {np.std(energies):.2f} MeV")
    print(f"Muon angle: mean = {np.rad2deg(np.mean(angles)):.2f}째, "
          f"std = {np.rad2deg(np.std(angles)):.2f}째")
    
    # Create figure with subplots
    fig = plt.figure(figsize=(15, 10))
    
    # 1. Momentum distribution
    ax1 = fig.add_subplot(221)
    mean, std = np.mean(momentum), np.std(momentum)
    ax1.hist(momentum, bins=50, histtype='bar', color='darkblue', label=f"\n$\\mathbf{{Mean}}$ = {mean:.2f}\n$\\mathbf{{Std}}$ = {std:.2f}")
    ax1.set_xlabel('Momentum (MeV/c)', fontsize=20)
    ax1.set_ylabel('Counts', fontsize=20)
    ax1.set_title('Muon Momentum Distribution - lab frame', fontsize=20)
    ax1.grid(True)
    leg = ax1.legend(loc='upper right', bbox_to_anchor=(0.95, 0.95), 
                      fontsize='small', title=f'$\\mathbf{{Summary}}$', title_fontsize='medium')
    leg.get_title().set_fontweight('bold')
    leg._legend_box.align = "center"
    
    # 2. Energy distribution
    ax2 = fig.add_subplot(222)
    mean, std = np.mean(energies), np.std(energies)
    ax2.hist(energies, bins=50, histtype='bar', color='brown', label=f"\n$\\mathbf{{Mean}}$ = {mean:.2f}\n$\\mathbf{{Std}}$ = {std:.2f}")
    ax2.set_xlabel('Energy (MeV)', fontsize=20)
    ax2.set_ylabel('Counts', fontsize=20)
    ax2.set_title('Muon Energy Distribution - lab frame', fontsize=20)
    ax2.grid(True)
    leg = ax2.legend(loc='upper right', bbox_to_anchor=(0.95, 0.95), 
                      fontsize='small', title=f'$\\mathbf{{Summary}}$', title_fontsize='medium')
    leg.get_title().set_fontweight('bold')
    leg._legend_box.align = "center"
    
    # 3. Angular distribution
    ax3 = fig.add_subplot(223)
    mean, std = np.rad2deg(np.mean(angles)), np.rad2deg(np.std(angles))
    ax3.hist(np.rad2deg(angles), bins=50, histtype='bar', color='darkgreen', label=f"\n$\\mathbf{{Mean}}$ = {mean:.2f}\n$\\mathbf{{Std}}$ = {std:.2f}")
    ax3.set_xlabel('Angle (degrees)', fontsize=20)
    ax3.set_ylabel('Counts', fontsize=20)
    ax3.set_title('Muon Angular Distribution - lab frame', fontsize=20)
    ax3.grid(True)
    leg = ax3.legend(loc='upper right', bbox_to_anchor=(0.95, 0.95), 
                      fontsize='small', title=f'$\\mathbf{{Summary}}$', title_fontsize='medium')
    leg.get_title().set_fontweight('bold')
    leg._legend_box.align = "center"
    
    # 4. Add momentum-energy correlation plot
    ax5 = fig.add_subplot(224)
    ax5.scatter(momentum, energies, color='black', alpha=0.1, s=1)
    ax5.set_xlabel('Momentum (MeV/c)', fontsize=20)
    ax5.set_ylabel('Energy (MeV)', fontsize=20)
    ax5.set_title('Muon Energy vs Momentum - lab frame', fontsize=20)
    ax5.grid(True)
    
    # Make plots tight and show
    plt.tight_layout()
    plt.show()
```

```python
analyze_results(muons_lab)
```

```python
import numpy as np
from Particle import Particle

class PionDecaySim:
    '''
    Simulates +  關+ + 館關 decay
    '''
    # Particle masses in MeV/c^2
    PION_MASS = 139.57  # + mass MeV
    MUON_MASS = 105.66  # 關+ mass MeV
    C = 3e8 # m/s
    
    def __init__(self, n_particles, pion_momentum_mean, pion_momentum_std):
        self.n_particles = n_particles
        self.pion_momentum_mean = pion_momentum_mean
        self.pion_momentum_std = pion_momentum_std
        
    def pion_decay_lab_frame(self):
        '''
        Generate distribution of pions resulting from decay in lab frame
        '''
        momentum = np.random.normal(self.pion_momentum_mean, self.pion_momentum_std, self.n_particles)
        cos_theta = np.random.uniform(-1, 1, self.n_particles)
        phi = np.random.uniform(0, 2*np.pi, self.n_particles)
        sin_theta = np.sqrt(1 - cos_theta**2)
        
        px = momentum*sin_theta*np.cos(phi)
        py = momentum*sin_theta*np.sin(phi)
        pz = momentum*cos_theta
        
        pions = []
        for i in range(self.n_particles):
            momentum_vec = np.array([px[i], py[i], pz[i]])
            energy = np.sqrt((self.PION_MASS*self.PION_MASS) + np.sum(momentum_vec**2))
            pions.append(Particle(self.PION_MASS, momentum_vec, energy))
        return pions
            
    def after_decay_particles_rest_frame(self):
        '''
        Generate muon and neutrino momenta in rest frame after pion decay
        '''
        # Two-body decay momentum in rest frame
        momentum = (self.PION_MASS**2 - self.MUON_MASS**2) / (2*self.PION_MASS)
        
        cos_theta = np.random.uniform(-1, 1, self.n_particles)
        phi = np.random.uniform(0, 2*np.pi, self.n_particles)
        sin_theta = np.sqrt(1 - cos_theta**2)
        
        px = momentum*sin_theta*np.cos(phi)
        py = momentum*sin_theta*np.sin(phi)
        pz = momentum*cos_theta
        
        muons_momentum, nus_momentum = [], []
        for i in range(self.n_particles):
            muon_mom = np.array([px[i], py[i], pz[i]])
            muons_momentum.append(muon_mom)
            nus_momentum.append(-muon_mom)
            
        return muons_momentum, nus_momentum
    
    def after_decay_particles_lab_frame(self, rest_momentum, mass, pion):
        '''
        Generate particle momentum in lab frame after pion decay using Lorentz boost
        '''
        # Debug prints
        print("Input parameters:")
        print(f"Rest momentum: {rest_momentum}")
        print(f"Mass: {mass}")
        print(f"Pion momentum: {pion.get_momentum()}")
        print(f"Pion energy: {pion.get_energy()}")
        
        # Calculate rest frame energy
        E_rest = np.sqrt(mass*mass + np.sum(rest_momentum**2))
        print(f"Rest frame energy: {E_rest}")
        
        # Get boost parameters
        p_pion = np.linalg.norm(pion.get_momentum())
        E_pion = pion.get_energy()
        beta = p_pion/E_pion  # magnitude of velocity
        gamma = E_pion/self.PION_MASS
        print(f"Beta: {beta}")
        print(f"Gamma: {gamma}")
        
        # Get boost direction
        boost_direction = pion.get_momentum() / p_pion
        print(f"Boost direction: {boost_direction}")
        
        # Get momentum components
        p_parallel = np.dot(rest_momentum, boost_direction)
        p_parallel_vec = p_parallel * boost_direction
        p_perp_vec = rest_momentum - p_parallel_vec
        print(f"Parallel momentum: {p_parallel}")
        print(f"Perpendicular momentum: {p_perp_vec}")
        
        # Lorentz boost
        p_lab_parallel = gamma * (p_parallel + beta * E_rest)
        p_lab = p_perp_vec + p_lab_parallel * boost_direction
        E_lab = gamma * (E_rest + beta * p_parallel)
        print(f"Lab frame momentum: {p_lab}")
        print(f"Lab frame energy: {E_lab}")
        
        # Check conservation
        invariant_mass_squared = E_lab*E_lab - np.sum(p_lab**2)
        print(f"Invariant mass squared: {invariant_mass_squared}")
        print(f"Expected mass squared: {mass*mass}")
        print(f"Difference: {abs(invariant_mass_squared - mass*mass)}")
        
        if abs(invariant_mass_squared - mass*mass) > 1e-6:  # Relaxed tolerance
            raise ValueError("Energy-momentum conservation violated")
            
        return Particle(mass, p_lab, E_lab)

# Helper function to verify kinematics
def verify_kinematics(particle):
    """Verify that E^2 - p^2 = m^2 for a given particle"""
    E = particle.energy
    p = particle.momentum
    m = particle.mass
    
    E_squared = E*E
    p_squared = np.sum(p**2)
    m_squared = m*m
    
    print(f"E^2: {E_squared}")
    print(f"p^2: {p_squared}")
    print(f"m^2: {m_squared}")
    print(f"E^2 - p^2 - m^2: {E_squared - p_squared - m_squared}")
    
    return abs(E_squared - p_squared - m_squared) < 1e-6
```
